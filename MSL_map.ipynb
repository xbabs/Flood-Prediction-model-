```python
import pandas as pd
import numpy as np
import folium
```

```python
map_data = pd.read_excel("data.xlsx")
map_data.head(4)
```

```python
map_data.isnull().sum()
```

```python
from pandas._libs.lib import map_infer
# Calculate the center latitude and longitude based on the mean of coordinates
center_latitude = map_data['LATITUDE'].mean()
center_longitude = map_data['LONGITUDE'].mean()

# Create a map centered at the calculated location
#m = folium.Map(location=[center_latitude, center_longitude], zoom_start=10)

# Create a map centered at a specific location (adjust latitude and longitude)
map_center = [center_latitude, center_longitude]
m = folium.Map(location=map_center, zoom_start=10)

# Loop through monitoring stations and add markers to the map
for index, row in map_data.iterrows():
    station_name = row['STATION']
    latitude = row['LATITUDE']
    longitude = row['LONGITUDE']
    mean_sea_level = row['MEAN_SEA_LEVEL']

    # Create a marker with a popup displaying mean sea level
    popup_text = f"Station: {station_name}<br>Mean Sea Level: {mean_sea_level:.2f}"
    folium.Marker(location=[latitude, longitude], popup=popup_text).add_to(m)

# Display the map
m.save('mean_sea_level_map.html')  # Save the map as an HTML file
```

```python
from IPython.display import HTML

# Read the content of the HTML file
with open('mean_sea_level_map.html', 'r') as file:
    map_html = file.read()

# Display the HTML content as an embedded map
HTML(map_html)
```
